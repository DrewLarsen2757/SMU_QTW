# 0. Load data

```r
df <- read.csv("Case2_FullDataFrame.csv", header = TRUE)
```

```r
# removing unecessary variable ("X" and  "sex")
df = df[,(names(df)!= "X" & names(df)!= "sex")]
```

```r
head(df)
```

# 2. DATA EXPLORATION

```r
str(df)
```

```r
summary(df)
```

```r
#convert year to a factor variable instead of  numerical variable
df$year<- factor(df$year) 
```

```r
str(df)
```

```r
### TODO  print NA total
```

```r
#remove na quick fix
df <- na.omit(df) 
```

```r
age = sapply(df, function(x) as.numeric(x[ , 'ag']))
```

```r
### box plot of age distribution  per year
# grouped boxplot
ggplot(df, aes(x=year, y=age)) + geom_boxplot()
```

The box plot distribution across year is a good indication to confirm that the average age of participant across year is consistent and it's between [30 to 35]

```r
hist(df$runTime, xlab = "Time(minutes)", main = "Distribution of time, Female Runners", 
    col = "blue", breaks = 10)
```

From the runTime distribution above we can see  that the runtime is normally distributed from 1999-2012. This means that  the expected female runtime is around 98 Min. 


```r
hist(df$age, xlab = "age", main = "Distribution of age across Female Runners", 
    col = "red", breaks = 10)
```

From the age distribution above we can see the participant's age goes from 10 to 70 and most of the participant are between 20-40. the data is skew to the right so we have some other participant form 40 to 60


```r
# A basic scatterplot with color depending on Species
ggplot(df, aes(x=age, y=runTime, color=year)) + 
    geom_point(size=6) 
```

The scatter plot above across year doesn't show any particular linear trend but show a big cluster which tell us year by year the most of the participant are between  age 25 and 50 and  runtime are between 75 and 125 min.

```r
# #dot correlation
data_num <- select_if(df, is.numeric)   
correlations <- cor(select_if(df, is.numeric))
corrplot(correlations, method="circle")
```

```r
the correlation plot between the age and runTime, confirmed on the idea of weak correlation betwenn the two variable
```

```r
ggplot(df, aes(year)) +
  geom_bar(fill = "#0073C2FF")
#   theme_pubclean()
```

As you can see from the plot above the number of participant increase every year, from 1999 to 2012 we have almost 4X the number of participant from 2500 to 10000. 

```r
# with linear trend
ggplot(df, aes(x=age, y=runTime)) +
  geom_point() +
  geom_smooth(method=lm , color="red", fill="#69b3a2", se=TRUE) 
```

From the plot above we can see there's not a clear linear trend or correlation between the age and the runtime. but confirm on the big cluster.
